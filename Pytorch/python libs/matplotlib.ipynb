# matplotlib

## matplotlib 中的子模块 pyplot
pyplot 是 matplotlib 中最常用的子模块，用于绘制散点图、柱状图、折线图、直方图、饼图等
一般来说 import 的基本方法就是：import matplotlib.pyplot as plt

```python
import matplotlib.pyplot as plt
import numpy as np
```

## plt 子图创建
子图的创建可以是一些简单创建，比如：

```python
fig = plt.figure(num=1,figsize=(4,4))
ax1 = fig.add_subplot(221)
#可从图中看到，我们的画布是分为2x2的区域
#我们可以通过add_subplot来添加子图，参数是一个三位数，分别代表行数，列数，子图的编号
ax1.plot([1,2,3,4],[1,2,3,4])
ax2 = fig.add_subplot(222)
ax2.plot([1,2,3,4],[2,2,3,4])
ax3 = fig.add_subplot(223)
ax3.plot([1,2,3,4],[1,2,2,4])
ax4 = fig.add_subplot(224)
ax4.plot([1,2,3,4],[1,2,3,3])

plt.show()
```

## matplotlib提供了两种最常用的绘图接口

1. 显式创建figure和axes，在上面调用绘图方法，也被称为OO模式（object-oriented style）

2. 依赖pyplot自动创建figure和axes，并绘图

### 第一种 OO 方式

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the same axes.
ax.plot(x, x**3, label='cubic')  # Plot more data on the same axes.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add legend (label of each plot)

plt.show()
```

### plt 自动创建 (使用较为简单)

```python
x = np.linspace(0, 2, 100)
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # Plot more data on the same axes.
plt.plot(x, x**3, label='cubic')  # Plot more data on the same axes.
plt.xlabel('x label')  # Add an x-label to the axes.
plt.ylabel('y label')  # Add a y-label to the axes.
plt.title("Simple Plot")  # Add a title to the axes.
plt.legend()  # Add legend (label of each plot)

plt.show()
```

## 通用绘图模板

由于 matplotlib 的知识点非常繁杂，在实际使用过程中也不可能将全部 API 都记住，很多时候都是边用边查。因此这里提供一个通用的绘图基础模板，任何复杂的图表几乎都可以基于这个模板骨架填充内容而成。初学者刚开始学习时只需要牢记这一模板就足以应对大部分简单图表的绘制，在学习过程中可以将这个模板模块化，了解每个模块在做什么，在绘制复杂图表时如何修改，填充对应的模块。

```python
import matplotlib as mpl
# step 1: 准备数据
x = np.linspace(0, 2, 100)
y = x ** 2

# step 2: 设置绘图样式（非必须，并且可以在绘制图像时进行设置
mpl.rc('lines', linewidth=4, linestyle='--', color='r')

# step 3: 绘制图像
plt.plot(x, y, label='linear')

# step 4: 添加标签，文字和图例
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

### Tips
在jupyter notebook中使用matplotlib时会发现，代码运行后自动打印出类似 <matplotlib.lines.Line2D at 0x23155916dc0> 这样一段话，这是因为 matplotlib 的绘图代码默认打印出最后一个对象。如果不想显示这句话，有以下三种方法:

1. 在代码块最后加一个分号 ';'(不美观)

2. 在代码块最后加一句plt.show()（相对推荐）

3. 在绘图时将绘图对象显式赋值给一个变量，如将plt.plot([1, 2, 3, 4]) 改成line =plt.plot([1, 2, 3, 4])
   多一个变量不知道怎么命名
