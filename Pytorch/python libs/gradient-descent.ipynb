# 简单的梯度下降算法实现

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x : (x - 1) ** 2 + 1
df = lambda x : 2 * (x - 1)

cur_x = 6
cur_y = f(cur_x)
learning_rate = 0.1
iterations = 100

all_x = [cur_x]
all_y = [cur_y]
```

## 循环过程

```python
for i in range(iterations):
    gradient = df(cur_x)
    cur_x -= learning_rate * gradient
    cur_y = f(cur_x)
    all_x.append(cur_x)
    all_y.append(cur_y)
```

## 绘图过程

```python
x = np.linspace(-5, 7, 100)
y = f(x)

plt.axis([-5, 7, 0, 50])
plt.plot(x, y)
print(all_x)
plt.scatter(all_x, all_y, c='r')

plt.show()
```
